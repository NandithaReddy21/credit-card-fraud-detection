```python
#Scale Time and Amount
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['scaled_amount'] = scaler.fit_transform(df[['Amount']])
df['scaled_time'] = scaler.fit_transform(df[['Time']])

df.head()
```

```python
df = df.drop(['Time', 'Amount'], axis = 1)

df.head()
```

Logistic Regression is distance-sensitive. Since Time and Amount have much larger magnitudes than the PCA features, scaling prevents them from dominating the optimization process.

```python
#Train-Test Split

X = df.drop('Class', axis = 1)
y = df['Class']

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2, #80% training data, 20% testing data
    random_state = 42,
    stratify = y 
)
```

```python
print(X_train.shape)
print(X_test.shape)

print(f"\nTraining : {y_train.value_counts()}")
print(f"\nTesting : {y_test.value_counts()}")
```

## First Model: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
#Evaluating the model

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy : ', accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)  
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

## Logistic Regression Results

The baseline Logistic Regression model achieved an accuracy of 99.91%.

However, due to the highly imbalanced nature of the dataset, accuracy alone is not a reliable metric.

The model achieved:
- Precision: 83%
- Recall: 64%
- F1-Score: 72%

The model successfully detected 63 fraudulent transactions but missed 35 fraud cases.

Since missing fraudulent transactions can lead to financial losses, improving recall will be an important objective in subsequent experiments.

SMOTE (Synthetic Minority Over-sampling Technique) is a technique used to handle imbalanced datasets. Instead of duplicating minority-class samples, it generates synthetic samples by interpolating between existing minority examples. In my fraud detection project, I applied SMOTE only on the training data to improve the model's ability to detect fraudulent transactions while avoiding data leakage.


The test set should represent unseen real-world data. Applying SMOTE to the test set would introduce synthetic samples and cause data leakage, leading to overly optimistic performance estimates.

```python
from imblearn.over_sampling import SMOTE 

smote = SMOTE(random_state = 42)

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
print(y_train.value_counts())
print()
print(y_train_smote.value_counts())
```

```python
#Train Logistic Regression on SMOTE Data

lr_smote = LogisticRegression(max_iter = 1000)

lr_smote.fit(X_train_smote, y_train_smote)
```

```python
y_pred_smote = lr_smote.predict(X_test)
```

```python
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_smote))

print('\nClassification Report:')
print(classification_report(y_test, y_pred_smote))
```

## Effect of SMOTE

Applying SMOTE significantly improved recall from 64% to 92%, reducing the number of missed fraud cases from 35 to 8.

However, precision decreased substantially due to a large increase in false positives. This highlights the tradeoff between fraud detection capability and false alarm rate in imbalanced classification problems.

Further experimentation with more powerful models such as Random Forest will be performed to achieve a better balance between precision and recall.

## Handling Class Imbalance using scale_pos_weight

The dataset is highly imbalanced, containing significantly fewer fraudulent transactions than normal transactions.

Instead of generating synthetic samples using SMOTE, XGBoost provides a built-in mechanism called `scale_pos_weight`.

The value is calculated as:

scale_pos_weight = Number of Non-Fraud Transactions / Number of Fraud Transactions

This increases the penalty for misclassifying fraudulent transactions during training, encouraging the model to pay more attention to the minority class.

This approach is commonly used in industry because it preserves the original data distribution and avoids introducing synthetic samples.

```python
neg = y_train.value_counts()[0]
pos = y_train.value_counts()[1]

scale_pos_weight = neg/pos

print(scale_pos_weight)
```

```python
xgb_weighted = XGBClassifier(
    random_state = 42,
    eval_metric = 'logloss',
    scale_pos_weight = scale_pos_weight
)

xgb_weighted.fit(X_train, y_train)
```

```python
y_pred_xgb_weighted = xgb_weighted.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb_weighted))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb_weighted))
```

## Cross Validation

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

StratifiedKFold

Splits the data into multiple folds while preserving the fraud/non-fraud ratio.

Without stratification, some folds might contain very few fraud cases.

```python
cv = StratifiedKFold(
    n_splits = 5, 
    shuffle = True,
    random_state = 42
)
```

```python
recall_scores = cross_val_score(
    xgb_weighted,
    X,
    y,
    cv = cv,
    scoring = 'recall',
    n_jobs = -1
)
```

```python
print('Recall scores: ', recall_scores)
print('Mean Recall:', recall_scores.mean())
print('Standard Deviation:', recall_scores.std())
```

## Cross Validation Results

To evaluate model robustness, 5-fold Stratified Cross Validation was performed using Recall as the evaluation metric.

StratifiedKFold was selected because the dataset is highly imbalanced and each fold preserves the original fraud-to-non-fraud ratio.

### Results

- Recall Scores: [0.828, 0.859, 0.837, 0.786, 0.816]
- Mean Recall: 82.5%
- Standard Deviation: 2.4%

### Interpretation

The model achieved a mean recall of 82.5% across five folds, indicating that it consistently detected the majority of fraudulent transactions.

The low standard deviation suggests that performance is stable across different subsets of the data and is not heavily dependent on a particular train-test split.

These results provide additional confidence that the model generalizes well to unseen data.

## ROC Curve + ROC_AUC

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_probs = xgb_weighted.predict_proba(X_test)[:,1]

auc = roc_auc_score(y_test, y_probs)

print('ROC-AUC Score :', auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_probs) #fpr = false positive rate = FP/(FP+TN), tpr = true positive rate = TP/(TP+FN)

plt.figure(figsize = (6,4))
plt.plot(fpr, tpr, label = f'Weighted XGBoost (AUC = {auc: .3f})')

plt.plot([0,1],[0,1], linestyle = '--', label = 'Random Guessing (AUC = 0.50)')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')

plt.legend()

plt.savefig('images/roc_curve.png', bbox_inches = 'tight')
plt.show()
```
